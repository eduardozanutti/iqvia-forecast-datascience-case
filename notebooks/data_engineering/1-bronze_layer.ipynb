```python
import pandas as pd
```

```python
df = pd.read_csv('../../data/raw/dataset-case-iqvia.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.columns = ['week_date','supplier_id','product_id','region','units_sold','product_attribute_1','product_attribute_2','product_attribute_3']
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['week_date'] = pd.to_datetime(df['week_date'], format='%Y-%m-%d')
df['supplier_id'] = df['supplier_id'].astype('object')
df['product_id'] = df['product_id'].astype('object')
df['units_sold'] = df['units_sold'].astype('int')
```

```python
df.dtypes
```

```python
df.to_parquet('../../data/bronze/iqvia-sales_data.parquet', index=False)
```
