# Bronze Layer — Raw Ingestion

**Objective:** Ingest the raw CSV file, enforce a consistent schema (column names + data types),
and persist the result as Parquet for downstream processing.

**Input :**   
**Output:** 

```python
import pandas as pd
```

```python
# ── 1. Load raw CSV ──────────────────────────────────────────────────────────
df = pd.read_csv('../../data/raw/dataset-case-iqvia.csv')
print(f"{len(df):,} rows  |  {df.shape[1]} columns")
df.head()
```

```python
# ── 2. Standardise column names ──────────────────────────────────────────────
df.columns = [
    'week_date', 'supplier_id', 'product_id', 'region',
    'units_sold', 'product_attribute_1', 'product_attribute_2', 'product_attribute_3',
]
```

```python
# ── 3. Enforce data types ────────────────────────────────────────────────────
df['week_date']   = pd.to_datetime(df['week_date'], format='%Y-%m-%d')
df['supplier_id'] = df['supplier_id'].astype('object')
df['product_id']  = df['product_id'].astype('object')
df['units_sold']  = df['units_sold'].astype('int')

df.dtypes
```

```python
# ── 4. Save to bronze ────────────────────────────────────────────────────────
df.to_parquet('../../data/bronze/iqvia-sales_data.parquet', index=False)
print('Saved: data/bronze/iqvia-sales_data.parquet')
```
